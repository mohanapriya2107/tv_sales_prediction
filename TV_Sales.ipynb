```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/advertising.csv")
df
```

```python
X=df[['TV']]
y=df['Sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
training_results=reg.predict(X_train)
training_results
```

```python
X_train=X_train.to_numpy().ravel()
```

```python
training_data=pd.DataFrame({"X_Train":X_train,
                            "y_train":y_train,
                            "y_pred":training_results})
training_data
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error
print("Training Performance")
print("---------------------")
print(f"Accuracy of Training Data:{r2_score(X_train,training_results)}")
print(f"Loss of Training Data:{root_mean_squared_error(X_train,training_results)}")
```
